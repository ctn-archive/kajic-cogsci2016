```python
from vector_similarities.compute_similarity import compute_similarity
import matplotlib.pyplot as pl
%matplotlib inline
```

### SVD-decomposed free norms

##### 128 dimensions

```python
st, se = compute_similarity('svd_5018w_16d')

pl.figure()
pl.plot(st.mean(axis=1), label='target', lw=2)
pl.plot(se.mean(axis=1), label='all words', lw=2)
pl.title('Average cues similarities to:')
pl.xlabel('RAT problem ID')

pl.legend()
```
